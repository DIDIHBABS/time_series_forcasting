```python
import pandas as pd 
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import r2_score
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
from sklearn.model_selection import TimeSeriesSplit


```

```python
data = pd.read_csv("Sunspots_database.csv", usecols=[1, 2])
```

```python
data.head()
```

```python
data. describe()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format= '%Y-%m-%d')
data['year'] = data['Date'].dt.year
df = data.groupby('year').sum()
```

```python
df = df.drop_duplicates(subset=['date_time'], keep=False)

```

```python
df.head()
```

```python

ax = df.plot()
ax.set_ylabel('Sunspots')
plt.show()
```

# ADF test
## is the data Sationary

```python
test_result=adfuller(df['Monthly Mean Total Sunspot Number'])
```

```python
def adf_test(sunspots):
    result = adfuller(sunspots)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    
    if result[1] <= 0.05:
        print(" strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary ")
    else:
        print("weak evidence against null hypothesis,indicating it is non-stationary ")

adf_test(df['Monthly Mean Total Sunspot Number'])
```

# ARMA Model 

## ACF and PACF plots

```python
plot_acf(df['Monthly Mean Total Sunspot Number'], lags=50)

plot_pacf(df['Monthly Mean Total Sunspot Number'], lags=50)

plt.show()
```

```python

# Forecast the first ARMA(1,1) model
model1 = ARIMA(list(df['Monthly Mean Total Sunspot Number']), order=(1,0,1))
res = model1.fit()
pred = res.predict()
print(r2_score(df, pred))

plt.plot(list(df['Monthly Mean Total Sunspot Number']))
plt.plot(pred)
plt.legend(['Actual Sunspots', 'Predicted Sunspots'])
plt.xlabel('Timesteps')
plt.show()
```

```python
ax = pd.Series(res.resid).hist()
ax.set_ylabel('Number of occurences')
ax.set_xlabel('Residual')
plt.show()
```
